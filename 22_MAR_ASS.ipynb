### Q1. Pearson correlation coefficient is a measure of the linear relationship between two variables. Suppose you have collected data on the amount of time students spend studying for an exam and their final exam scores. Calculate the Pearson correlation coefficient between these two variables and interpret the result.

To calculate the Pearson correlation coefficient between the amount of time students spend studying for an exam and their final exam scores, you would need a dataset that includes both variables for each student. Once you have the dataset, you can follow these steps to calculate the Pearson correlation coefficient:

* Calculate the mean (average) of the time spent studying and the mean of the exam scores.
* Calculate the difference between each student's time spent studying and the mean of the time spent studying.
* Calculate the difference between each student's exam score and the mean of the exam scores.
* Multiply the corresponding differences obtained in steps 2 and 3 for each student and sum them up.
* Square each difference obtained in steps 2 and 3, sum them up separately, and calculate the square roots of both sums.
* Divide the sum calculated in step 4 by the product of the square roots calculated in step 5.
* The result of step 6 is the Pearson correlation coefficient.
